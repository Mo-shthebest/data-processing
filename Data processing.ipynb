```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/WiDS Dataton.csv')
print(data.shape)
data.head()
data.info()
```

```python
data.columns
```

```python
data.drop(['id'],axis=1,inplace=True)
data.head()
```

```python
print('Number of dublicated:', data.duplicated().sum())
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

```python
data.isnull().sum().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(20,10))
data.isnull().sum().sort_values(ascending=False).head(6).plot(kind='bar')
plt.title('Missing Values',fontsize=20)
plt.ylim(0,data.shape[0])
plt.show()
```

```python
data.drop(['direction_max_wind_speed',
       'direction_peak_wind_speed', 'max_wind_speed','days_with_fog','energy_star_rating'],axis=1,inplace=True)
data.shape
```
